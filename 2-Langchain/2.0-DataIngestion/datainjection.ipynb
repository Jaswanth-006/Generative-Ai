```python
from langchain_community.document_loaders import WebBaseLoader

import bs4
loader = WebBaseLoader(web_path=("https://python.langchain.com/api_reference/community/document_loaders/langchain_community.document_loaders.text.TextLoader.html",),
                       bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                           class_=("post-title","post-content","post-header")
                       ))
                       )
docs = loader.load()
docs
```
